```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
# read files from ./rawdata
files = os.listdir('./rawdata')
# emerge all file into one dataframe
data = pd.DataFrame()
for file in files:
    data = pd.concat([data, pd.read_excel('./rawdata/' + file)])
```

```python
data[data < -1]=np.nan
```
